```python
import requests
import json
import pandas as pd

# importação dos dados_hospedagem.
url = "https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_hospedagem.json"
resposta = requests.get(url)
# Convertendo o JSON para DataFrame
data_hospedagem  = pd.DataFrame(resposta.json())
```

```python
#visualizando nosso dataframe importado &
# Exportar para JSON - caso vocês queiram ficar com o arquivo na máquina.
data_hospedagem.to_json("data_hospedagem.json", orient="records", indent=4)
data_hospedagem.head()
```


### Quando um arquivo JSON possui dados aninhados (com vários níveis de aninhamento), o método json_normalize() é utilizado para transformar esses dados em um formato de tabela com múltiplas colunas.

Perceba que os dados estão em formato de discionário e não em formato de dataframe. vamos aplicar, então, a normalize


```python
data_hospedagem = pd.json_normalize(data_hospedagem['info_moveis'])
data_hospedagem
```

```python
colunas = list(data_hospedagem.columns)
colunas
```

## Outras formas de tratar um texto:



```python
# Regex
data_hospedagem['descricao_local'][3170]
# veja que tem hífen, apóstrofo, ponto de exclamação
```

```python
data_hospedagem['descricao_local'] = data_hospedagem['descricao_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex= True) # selecionar todos os carcteres com exceção daqueles dentro do []
```

```python
data_hospedagem['descricao_local'] = data_hospedagem['descricao_local'].str.replace('(?<!\w)-(?!\w)',' ', regex= True) # verifica se há caracter depois do hífen
data_hospedagem.head()
```

Regex (ou expressão regular) é uma sequência de caracteres que define um padrão de busca em um texto. Ele é uma ferramenta que permite encontrar, substituir e manipular padrões de texto. O regex é utilizado em diferentes áreas, incluindo programação, ciência de dados e processamento de texto.

Faça a transformação dos dados na coluna preco do conjunto de dados moveis_disponiveis.json para o tipo numérico float64. Lembrando que, antes de fazer isso, você precisa preencher os valores vazios na coluna por um valor. Uma sugestão: substitua pela string '0.0'.

```python
data_imoveis.head()
```

```python
# Substituindo 'None' (string) por '0.0' e valores None reais por 0.0
data_imoveis['preco'] = data_imoveis['preco'].replace('None', '0.0').fillna('0.0')
data_imoveis.head()
```

```python
# Depois, vamos transformar a coluna  para o tipo numérico com o astype.
data_imoveis['preco'] = data_imoveis['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
data_imoveis['preco'] = data_imoveis['preco'].astype(np.float64)
data_imoveis['preco']
```
